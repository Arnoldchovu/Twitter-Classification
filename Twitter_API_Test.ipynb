```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from nltk.corpus import stopwords
```

```python
query = "(from:Safaricom_Care) until:2022-12-14 since:2010-01-01"
tweets = []
limit = 10000

for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.username, tweet.content, tweet.likeCount])
```

```python
# Creating a dataframe with the column names
df = pd.DataFrame(tweets, columns=['Data', 'User', 'Tweet', 'LikeCount'])
```

```python

# create a class that can be used to label tweets
class TweetLabeler:
  def __init__(self, tweet):
    self.tweet = tweet
  
label_mapping = {
  'Mpesa': ['pesa', 'funds', 'great'],
  'Fibre': ['fibre', 'faiba', 'terrible'],
  'Connection': ['signal', 'connection', 'network'],
  'Mshwari': ['mswari']
}

# create a list of labels for the tweets
labels = []

# iterate over the tweets and create a label for each one
for tweet in tweets:
  label = 'Unknown'
  for key, value in label_mapping.items():
    for word in value:
      if word in tweet:
        label = key
        break
    if label != 'Unknown':
      break
  labels.append(label)

```
